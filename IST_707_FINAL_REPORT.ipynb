# Introduction

Chronic absenteeism in educational institutions is a critical issue that can significantly affect student outcomes. The outcomes include missing early learning milestones, loss of productivity and social interaction, disruption of fundamental routines and rhythms, and, further, increased involvement in the criminal justice system (Gottfried, 2014; McCluskey et al., 2004).

Identifying at-risk students before they fall into patterns of frequent absences can enable timely interventions. In this study, we utilize the "Monitoring the Future" dataset, which is renowned for its comprehensive data on the behaviors, attitudes, and values of American secondary school students. By applying machine learning techniques, we aim to predict patterns of absenteeism - focusing on truancy - and explore the factors that contribute to chronic absenteeism. Our approach leverages a range of predictive models to analyze the dataset, aiming to provide educators and policymakers with tools to proactively address absenteeism and enhance student engagement.

# Section 2 Model Experiment

We use the following supervised learning algorithms: random forest, gradient boosting, SGD classifier, three different shapes of neural network, SVC, Voting classifier, KNN, and Adaboost.

### Section 2.1 Random Forest

We test different numbers of trees (n_estimators) in a Random Forest model to see how they affect accuracy and recall. A plot was generated to show these results. Then, we used Grid search to tune the model. Using the optimal parameters for Random Forest, the accuracy was 0.989, with Recall of 0.989, and F1 score of 0.987.

```python
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, recall_score
import matplotlib.pyplot as plt

# Initialize lists to store results
n_estimators_values = list(range(1, 21))
accuracies = []
recalls = []
best_n_estimators = None
best_accuracy = 0
best_recall = 0

# Perform experiment for different values of n_estimators
for n in n_estimators_values:
    rf_clf = RandomForestClassifier(n_estimators=n, random_state=42)

    rf_clf.fit(X_train, y_train)

    accuracy = rf_clf.score(X_test, y_test)
    accuracies.append(accuracy)

    y_pred = rf_clf.predict(X_test)
    recall = recall_score(y_test, y_pred,average='weighted')
    recalls.append(recall)

    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_recall = recall
        best_n_estimators = n

# Plot the results
plt.figure(figsize=(6, 4))
plt.plot(n_estimators_values, accuracies, marker='o', label='Accuracy')
plt.plot(n_estimators_values, recalls, marker='x', label='Recall')
plt.xlabel('Number of Trees (n_estimators)')
plt.ylabel('Score')
plt.title('Impact of Number of Trees on Random Forest Performance')
plt.grid(True)
plt.xticks(np.arange(0, 21, step=10))
plt.legend()
plt.show()

print(f"Best n_estimators: {best_n_estimators} (Accuracy: {best_accuracy}, Recall: {best_recall})")
```

As the target variable distribution was heavily right skewed (e.g., majority of the students did not miss class), average = 'weighted' was used for recall. The weighted recall takes into account the imbalance by weighting the recall of each class by the number of true instances for each class.

While this was an intentional decision, this made the accuracy and recall appear the same on the graph.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, recall_score

# Define hyperparameters and depths to search
n_estimators_values = list(range(1, 21))
max_depth_values = list(range(1, 11))

# Initialize lists to store results
results = []

# Perform grid search for different values of n_estimators and max_depth
for n_estimators in n_estimators_values:
    for max_depth in max_depth_values:
        # Initialize and train the model
        rf_clf = RandomForestClassifier(n_estimators=n_estimators, max_depth=max_depth, random_state=42)
        rf_clf.fit(X_train, y_train)

        # Predict on test set
        y_pred = rf_clf.predict(X_test)

        # Calculate accuracy and recall
        accuracy = accuracy_score(y_test, y_pred)
        recall = recall_score(y_test, y_pred, average='weighted')

        # Store the results
        results.append((n_estimators, max_depth, accuracy, recall))

# Convert results to numpy array
results = np.array(results)

# Find the row with the highest accuracy and recall
best_accuracy_index = np.argmax(results[:, 2])
best_recall_index = np.argmax(results[:, 3])

# Extract best parameters
best_accuracy_n_estimators = results[best_accuracy_index, 0]
best_accuracy_max_depth = results[best_accuracy_index, 1]
best_recall_n_estimators = results[best_recall_index, 0]
best_recall_max_depth = results[best_recall_index, 1]

# Print best parameters
print("Best parameters for Accuracy - n_estimators:", best_accuracy_n_estimators)
print("Best parameters for Accuracy - max_depth:", best_accuracy_max_depth)
print("Best parameters for Recall - n_estimators:", best_recall_n_estimators)
print("Best parameters for Recall - max_depth:", best_recall_max_depth)

# Extract values for plotting
n_estimators_values = results[:, 0]
max_depth_values = results[:, 1]
accuracy_values = results[:, 2]
recall_values = results[:, 3]

# Plotting the 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(n_estimators_values, max_depth_values, accuracy_values, c='blue', marker='o', label='Accuracy')
ax.scatter(n_estimators_values, max_depth_values, recall_values, c='red', marker='x', label='Recall')
ax.set_xlabel('Number of Estimators')
ax.set_ylabel('Max Depth')
ax.set_zlabel('Score')
ax.set_title('Grid Search Results')
ax.legend()
plt.show()
```

We plotted 3D scatter plots to search for the parameters for which the model has the highest accuracy and recall. From the results, the model performs best when the number of decision trees (n_estimators) is 7 and the maximum depth of each tree (max_depth) is 10, which maximizes both accuracy and good performance in terms of recall when performing the grid search for random forest classifiers.

As for how the accuracy and recall appearing the same, this was because average = 'weighted' was used for recall as mentioned above.

```python
## USING CROSS-VALIDATION

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

# Initialize the RandomForestClassifier correctly
rf_clf = RandomForestClassifier(n_estimators=100, random_state=42)

# Perform 5-fold cross-validation for accuracy
cv_accuracy = cross_val_score(rf_clf, X_train, y_train, cv=5, scoring='accuracy')
print("CV Accuracy Scores:", cv_accuracy)
print("Average CV Accuracy:", cv_accuracy.mean())

# Perform 5-fold cross-validation for recall
cv_recall = cross_val_score(rf_clf, X_train, y_train, cv=5, scoring='recall_weighted')
print("CV Recall Scores:", cv_recall)
print("Average CV Recall:", cv_recall.mean())

# Perform 5-fold cross-validation for F1 score
cv_f1 = cross_val_score(rf_clf, X_train, y_train, cv=5, scoring='f1_weighted')
print("CV F1 Scores:", cv_f1)
print("Average CV F1 Score:", cv_f1.mean())
```

After testing the random forest classifier using the optimal parameters to prevent overfitting, we got its accuracy as 0.989, recall as 0.989, and F1 score as 0.987. Cross-validation scores were also very stable each fold, showing that the model is reliable and not overfitted.

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import make_scorer, f1_score, accuracy_score, recall_score

# Define the model and parameters
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('clf', AdaBoostClassifier())
])
params = {'clf__n_estimators': [50, 100, 150]}
scoring = {'F1 Score': make_scorer(f1_score, average='weighted'),
           'Accuracy': make_scorer(accuracy_score),
           'Recall': make_scorer(recall_score, average='weighted')}

# Grid Search with refit set to 'F1 Score' to use the best model for predictions
grid_search = GridSearchCV(pipeline, param_grid=params, cv=5, scoring=scoring, refit='F1 Score', return_train_score=True)
grid_search.fit(X_train, y_train)

# Get the best model after fitting
best_model = grid_search.best_estimator_

# Evaluate on the test set using the best estimator
y_pred = best_model.predict(X_test)
test_f1 = f1_score(y_test, y_pred, average='weighted')
test_accuracy = accuracy_score(y_test, y_pred)
test_recall = recall_score(y_test, y_pred, average='weighted')

f1_adab=grid_search.cv_results_['mean_test_F1 Score'].max()

# Best scores from cross-validation
print("Cross-Validation Best Scores:")
print(f"Best CV F1 Score: {grid_search.cv_results_['mean_test_F1 Score'].max()}")
print(f"Best CV Accuracy: {grid_search.cv_results_['mean_test_Accuracy'].max()}")
print(f"Best CV Recall: {grid_search.cv_results_['mean_test_Recall'].max()}")

# Print test data results
print("Test Data Results:")
print(f"Test Data F1 Score: {test_f1}")
print(f"Test Data Accuracy: {test_accuracy}")
print(f"Test Data Recall: {test_recall}")

# Print train scores from CV to inspect overfitting
print("Train Scores from Cross-Validation:")
for score_name in ['mean_train_F1 Score', 'mean_train_Accuracy', 'mean_train_Recall']:
    mean_score = grid_search.cv_results_[score_name].max()
    print(f"{score_name}: {mean_score}")
```

The performance for the AdaBoost classifer model is as follows: accuracy is 0.982, recall is 0.982, and F1 score is 0.975. The recall score is the same as accuracy as we used average="weighted" taking into considering of the distribution of our target variable. Cross-validation on each score also indicates that the model is robust and not overfitted.

### Section 3.1 Random Forest

We test different numbers of trees (n_estimators) in a Random Forest model to see how they affect accuracy and recall. A plot was generated to show these results. Then, we used Grid search to tune the model. Using the optimal parameters for Random Forest, the accuracy was 0.988, with Recall of 0.988, and F1 score of 0.985.

```python
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, recall_score
import matplotlib.pyplot as plt

# Initialize lists to store results
n_estimators_values = list(range(1, 21))
accuracies = []
recalls = []
best_n_estimators = None
best_accuracy = 0
best_recall = 0

# Perform experiment for different values of n_estimators
for n in n_estimators_values:
    rf_clf = RandomForestClassifier(n_estimators=n, random_state=42)

    rf_clf.fit(X_train, y_train)

    accuracy = rf_clf.score(X_test, y_test)
    accuracies.append(accuracy)

    y_pred = rf_clf.predict(X_test)
    recall = recall_score(y_test, y_pred, average='weighted')
    recalls.append(recall)

    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_recall = recall
        best_n_estimators = n

# Plot the results
plt.figure(figsize=(6, 4))
plt.plot(n_estimators_values, accuracies, marker='o', label='Accuracy')
plt.plot(n_estimators_values, recalls, marker='x', label='Recall')
plt.xlabel('Number of Trees (n_estimators)')
plt.ylabel('Score')
plt.title('Impact of Number of Trees on Random Forest Performance')
plt.grid(True)
plt.xticks(np.arange(0, 21, step=10))
plt.legend()
plt.show()

print(f"Best n_estimators: {best_n_estimators} (Accuracy: {best_accuracy}, Recall: {best_recall})")
```

As the target variable distribution was heavily right skewed (e.g., majority of the students did not miss class), average = 'weighted' was used for recall. The weighted recall takes into account the imbalance by weighting the recall of each class by the number of true instances for each class.

While this was an intentional decision, this made the accuracy and recall appear the same on the graph.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, recall_score

# Define hyperparameters and depths to search
n_estimators_values = list(range(1, 21))
max_depth_values = list(range(1, 11))

# Initialize lists to store results
results = []

# Perform grid search for different values of n_estimators and max_depth
for n_estimators in n_estimators_values:
    for max_depth in max_depth_values:
        # Initialize and train the model
        rf_clf = RandomForestClassifier(n_estimators=n_estimators, max_depth=max_depth, random_state=42)
        rf_clf.fit(X_train, y_train)

        # Predict on test set
        y_pred = rf_clf.predict(X_test)

        # Calculate accuracy and recall
        accuracy = accuracy_score(y_test, y_pred)
        recall = recall_score(y_test, y_pred, average='weighted')

        # Store the results
        results.append((n_estimators, max_depth, accuracy, recall))

# Convert results to numpy array
results = np.array(results)

# Find the row with the highest accuracy and recall
best_accuracy_index = np.argmax(results[:, 2])
best_recall_index = np.argmax(results[:, 3])

# Extract best parameters
best_accuracy_n_estimators = results[best_accuracy_index, 0]
best_accuracy_max_depth = results[best_accuracy_index, 1]
best_recall_n_estimators = results[best_recall_index, 0]
best_recall_max_depth = results[best_recall_index, 1]

# Print best parameters
print("Best parameters for Accuracy - n_estimators:", best_accuracy_n_estimators)
print("Best parameters for Accuracy - max_depth:", best_accuracy_max_depth)
print("Best parameters for Recall - n_estimators:", best_recall_n_estimators)
print("Best parameters for Recall - max_depth:", best_recall_max_depth)

# Extract values for plotting
n_estimators_values = results[:, 0]
max_depth_values = results[:, 1]
accuracy_values = results[:, 2]
recall_values = results[:, 3]

# Plotting the 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(n_estimators_values, max_depth_values, accuracy_values, c='blue', marker='o', label='Accuracy')
ax.scatter(n_estimators_values, max_depth_values, recall_values, c='red', marker='x', label='Recall')
ax.set_xlabel('Number of Estimators')
ax.set_ylabel('Max Depth')
ax.set_zlabel('Score')
ax.set_title('Grid Search Results')
ax.legend()
plt.show()
```

```python
#############
#Experiments
#############

import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, recall_score
import matplotlib.pyplot as plt

# Initialize lists to store results
n_estimators_values = list(range(1, 21))
accuracies = []
recalls = []
best_n_estimators = None
best_accuracy = 0
best_recall = 0

# Perform experiment for different values of n_estimators

for n in n_estimators_values:
    rf_clf = RandomForestClassifier(n_estimators=n, random_state=42)

    rf_clf.fit(SX_train, Sy_train)

    accuracy = rf_clf.score(SX_test, Sy_test)
    accuracies.append(accuracy)

    y_pred = rf_clf.predict(SX_test)
    recall = recall_score(Sy_test, y_pred, average='weighted')
    recalls.append(recall)

    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_recall = recall
        best_n_estimators = n

# Plot the results
plt.figure(figsize=(6, 4))
plt.plot(n_estimators_values, accuracies, marker='o', label='Accuracy')
plt.plot(n_estimators_values, recalls, marker='x', label='Recall')
plt.xlabel('Number of Trees (n_estimators)')
plt.ylabel('Score')
plt.title('Impact of Number of Trees on Random Forest Performance')
plt.grid(True)
plt.xticks(np.arange(0, 21, step=10))
plt.legend()
plt.show()

print(f"Best n_estimators: {best_n_estimators} (Accuracy: {best_accuracy}, Recall: {best_recall})")
```

```python
import numpy as np
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, recall_score
import matplotlib.pyplot as plt

# Initialize lists to store results
n_estimators_values = list(range(1, 11))
accuracies = []
recalls = []
best_n_estimators = None
best_accuracy = 0
best_recall = 0

# Perform experiment for different values of n_estimators
for n in n_estimators_values:
    gb_clf = GradientBoostingClassifier(n_estimators=n, learning_rate=0.01, random_state=42)

    gb_clf.fit(SX_train, Sy_train)

    accuracy = gb_clf.score(SX_test, Sy_test)
    accuracies.append(accuracy)

    y_pred = gb_clf.predict(SX_test)
    recall = recall_score(Sy_test, y_pred, average='weighted')
    recalls.append(recall)

    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_recall = recall
        best_n_estimators = n

# Plot the results
plt.figure(figsize=(6, 4))
plt.plot(n_estimators_values, accuracies, marker='o', label='Accuracy')
plt.plot(n_estimators_values, recalls, marker='s', label='Recall')
plt.xlabel('Number of Estimators (n_estimators)')
plt.ylabel('Metric')
plt.title('Impact of Number of Estimators on Gradient Boosting Performance')
plt.grid(True)
plt.xticks(np.arange(0, 11, step=10))
plt.legend()
plt.show()

print(f"Best n_estimators: {best_n_estimators} (Accuracy: {best_accuracy}, Recall: {best_recall})")
```
